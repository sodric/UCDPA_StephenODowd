```python
# Now I want to include assists to the top goal scorers but weight them lower than scoring goals so 10 points a goal nd 5 points an assist. Then visualise this. 
```

```python
pl_players["Goal Threat"] = (pl_players["Goals"] *10) + (pl_players["Assists"]* 5)
```

```python
pl_players.nlargest(10, 'Goal Threat')
```

```python
Attack = pl_players.nlargest(10, 'Goal Threat')
```

```python
Name = (Attack["Name"]) 
Attack_threat = Attack["Goal Threat"]


plt.barh(Name,Attack_threat)
plt.title('Figure 8: Top 10 Attacking Premier League Players')
plt.ylabel('')
plt.xlabel('Goal Threat Score')
plt.show()
```

```python
# This inlcudes penalties which have a much higher change of resulting in a goal then say a shot from open play. Idenifty non-penalty goals & use this for weightings instead.
```

```python
pl_players["Non-Penalty Goals"] = (pl_players["Goals"]) - (pl_players["Penalty_Goals"])
```

```python
pl_players["Goal Threat Non Penalty"] = (pl_players["Non-Penalty Goals"] *10) + (pl_players["Assists"]* 5)
```

```python
pl_players.nlargest(10, 'Goal Threat Non Penalty')
```

```python
Attack_nonpen=pl_players.nlargest(10, 'Goal Threat Non Penalty')
```

```python
Name = (Attack_nonpen["Name"]) 
Attacknonpen_threat = Attack_nonpen["Goal Threat Non Penalty"]


plt.barh(Name,Attacknonpen_threat)
plt.title('Figure 9: Top 10 Attacking Premier League Players (excl. penalties)')
plt.ylabel('')
plt.xlabel('Goal Threat Score (excl. penalties)')
plt.show()
```
